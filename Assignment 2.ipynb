# Translation
Here original image will be translated using cv2.warpAffine 

```python
img = cv2.imread("C:/Users/HP/Desktop/download.jpg")
```

Original Image is shown as 

```python
plt.imshow(img[:,:,::-1])
plt.show()
```

Below, scaling is done with a shift of (100,50)

```python
import numpy as np
rows,cols, y = img.shape

M = np.float32([[1,0,100],[0,1,50]])
dst = cv2.warpAffine(img,M,(cols,rows))

cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Shift of (50,200)

```python
import numpy as np
rows,cols, y = img.shape

M = np.float32([[1,0,50],[0,1,200]])
dst = cv2.warpAffine(img,M,(cols,rows))

cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Rotation 
Now, the image can be rotated using cv2.getRotationMatrix2D

```python
img = cv2.imread("C:/Users/HP/Desktop/download.jpg")
```

Original Image:

```python
plt.imshow(img[:,:,::-1])
plt.show()
```

```python
rows,cols,y = img.shape

M = cv2.getRotationMatrix2D((cols/2,rows/2),90,1)
dst = cv2.warpAffine(img,M,(cols,rows))
```

```python
plt.imshow(dst[:,:,::-1])
plt.show()
```

Image is been rotated by 90 degrees 

```python
rows,cols,y = img.shape

M = cv2.getRotationMatrix2D((cols/2,rows/2),180,1)
dst = cv2.warpAffine(img,M,(cols,rows))
```

```python
plt.imshow(dst[:,:,::-1])
plt.show()
```

Image is been rotated by 180 degrees
